Connect to MA API server using grpc

```python
import grpc
from michelangelo.gen.api.v2.project_pb2 import Project
from michelangelo.gen.api.v2.project_svc_pb2 import ListProjectRequest, CreateProjectRequest, DeleteProjectRequest
from michelangelo.gen.api.v2.project_svc_pb2_grpc import ProjectServiceStub

metadata = (("rpc-service","ma-apiserver"), ("rpc-caller", "demo-client"), ("rpc-encoding", "proto"), ("context-ttl-ms", "2000"))
channel = grpc.insecure_channel("localhost:8080")
project_service = ProjectServiceStub(channel)
```

List projects

```python
list_project_request = ListProjectRequest(namespace="default", list_options={})
project_service.ListProject(list_project_request, metadata=metadata, timeout=1000)
```

Create a new project

```python
proj = Project()
proj.metadata.namespace = "default"
proj.metadata.name = "demo-project"
proj.spec.tier = 2
proj.spec.description = "demo project"
proj.spec.owner.owning_team = "8D8AC610-566D-4EF0-9C22-186B2A5ED793"
proj.spec.git_repo = "https://github.com/michelangelo-ai/michelangelo"
proj.spec.root_dir = "/demo-project"

create_project_request = CreateProjectRequest(project=proj)
project_service.CreateProject(create_project_request, metadata=metadata, timeout=1000)
```

List

```python
list_project_request = ListProjectRequest(namespace="default", list_options={})
project_service.ListProject(list_project_request, metadata=metadata, timeout=1000)
```

Delete the project

```python
delete_project_request = DeleteProjectRequest(namespace="default", name="demo-project")
project_service.DeleteProject(delete_project_request, metadata=metadata, timeout=1000)
```
